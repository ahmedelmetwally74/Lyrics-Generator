```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/kaggle/input/genius-song-lyrics-with-language-information/song_lyrics.csv")
data.head()
```

# Prepare English Dataset

```python
en = data[data['language'] == 'en']
en
```

```python
new_en = en[['artist','title' , 'lyrics']]
```

```python
en2= pd.read_csv('/kaggle/input/songs-dataset/Songs.csv')
en2
```

```python
new_en2 = en2.rename(columns={'Artist':'artist','Title': 'title', 'Lyrics': 'lyrics'})
```

```python
# Concatenate the DataFrames
english_df = pd.concat([new_en, new_en2], ignore_index=True)
english_df.head()
```

```python
english_df.shape
```

```python
english_df.to_csv('/kaggle/working/english_lyrics_df.csv', index=False)
```

# Prepare Arabic Dataset

```python
ar = data[data['language'] == 'ar']
ar
```

```python
ar['tag'].unique()
```

```python
# List of tags to display samples for
tags = ['rap', 'misc', 'rock', 'pop', 'rb', 'country']

# Loop through each tag and display some samples
for tag in tags:
    tag_df = ar[ar['tag'] == tag]
    if not tag_df.empty:
        print(f"Sample for tag '{tag}':")
        print(tag_df.sample(n=min(len(tag_df), 5)))
        print("\n")
    else:
        print(f"No data for tag '{tag}'\n")
```

```python
# filter the dataset and keep "pop", "rock", "rb"
tags_to_keep = ['pop', 'rock', 'rb']
new_ar = ar[ar['tag'].isin(tags_to_keep)]
new_ar['tag'].unique()
```

```python
new_ar.head()
```

```python
ar2 = pd.read_csv('/kaggle/input/songs-dataset/arabicLyrics.csv')
ar2.head()
```

```python
ar2['SongDialect'].unique()
```

```python
# Assuming ar2 is your DataFrame already filtered for 'ar' language
SongDialects = ['Meghribi', 'Gulf', 'Iraqi', 'Sudan', 'Egyptian', 'Levantine']

# Loop through each dialect and display some samples
for dialect in SongDialects:
    SongDialect_df = ar2[ar2['SongDialect'] == dialect]
    if not SongDialect_df.empty:
        print(f"Sample for SongDialect '{dialect}':")
        print(SongDialect_df.sample(n=min(len(SongDialect_df), 5)))
        print("\n")
    else:
        print(f"No data for SongDialect '{dialect}'\n")
```

```python
# filter the dataset and keep "pop", "rock", "rb"
SongDialect_to_keep = ['Egyptian', 'Levantine','Sudan']
new_ar2 = ar2[ar2['SongDialect'].isin(SongDialect_to_keep)]
new_ar2['SongDialect'].unique()
```

```python
new_ar2.head()
```

```python
new_ar.columns, new_ar2.columns
```

```python
"""
new_ar suggested important columns for our task: 
- title
- lyrics
new_ar2 suggested important columns for our task:
- songTitle
- Lyrics
"""
```

# Merge both to have one Arabic dataset

```python
new_ar = new_ar[['title', 'lyrics']]
new_ar2 = new_ar2[['SongTitle', 'Lyrics']]
new_ar2 = new_ar2.rename(columns={'SongTitle': 'title', 'Lyrics': 'lyrics'})
```
